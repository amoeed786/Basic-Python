```python
#Import the libraries
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the Data Set
iris=load_iris()

#Convert the Data Set into Data Frame
df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
print(df)
```

```python
#First 5 rows of Data Frame
print(df.head())
#Last 5 Rows of Data Frame
print()
print(df.tail())
```

```python
#Summary of Data Frame
print(df.describe())
```

```python
#Sum of te null values in each column of Data Frame
print(df.isnull().sum())
```

```python
#Types of columns of Data Frame
print(df.dtypes)
```

```python
#Inserting a new column of 'Sepal Area' which is the product of 'Sepal Width' and 'Sepal Width'
df['sepal area']=df['sepal length (cm)'] * df['sepal width (cm)']
df.head()
```

```python
# Add the target to the DataFrame
df['species'] = iris.target_names[iris.target]
df.head()
```

```python
# Filter the DataFrame for 'setosa' species 
setosadf=df[df['species']=='setosa']
setosadf.head()
```

```python
#Sorting the data frame by petal length in descending order
df=df.sort_values(by='petal length (cm)',ascending=False)
df.head()
```

```python
#Filterng the data frame with petal width greater than 1.5 cm
filter=df[df['petal width (cm)'] > 1.5]
filter.head()
```

```python
# Calculate the mean sepal width for each species
mean_sepal=df.groupby('species')['sepal width (cm)'].mean()
mean_sepal
```

```python
# Plotting the histogram
plt.figure(figsize=(8,10))
plt.hist(df['sepal length (cm)'], bins=20, color='yellow', edgecolor='black')
# Adding titles and labels
plt.title('Histogram of Sepal Length Distribution', fontsize=15)
plt.xlabel('Sepal Length', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# Create a dictionary to map species to colors
color_map = {'setosa': 'yellow', 'versicolor': 'red', 'virginica': 'green'}

plt.figure(figsize=(8,10))

# Loop through each species and plot scatter points with respective color
for species , color in color_map.items():
    spec = df[df['species']==species]
    plt.scatter(spec['sepal length (cm)'], spec['sepal width (cm)'],
                color=color, label=species, alpha=0.9, edgecolors='black')
# Add labels and title and legend
plt.title('Relationship between Sepal Length and Sepal Width by Species', fontsize=14)
plt.xlabel('Sepal Length', fontsize=12)
plt.ylabel('Sepal Width', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a boxplot
plt.figure(figsize=(8, 6))
# Plot the boxplot
df.boxplot(column='petal length (cm)', by='species', grid=False)
# Customize the plot with title and labels
plt.title('Distribution of Petal Length for Each Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
plt.show()
```

```python
# Create a pairplot colored by species
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='species')
# Add title to the pairplot
plt.suptitle('Pairplot of Iris Dataset Variables Colored by Species',fontsize=15,y=1.05)
plt.show()
```

```python
'''Plot 1: 
The histogram of 'sepal_length' for the entire Iris dataset shows that the majority of sepal lengths fall within the range of approximately
4.5 to 7.5 cm.
There is a peak around the range of 5.5 to 6.0 cm, indicating a common sepal length among the iris samples.
The distribution of sepal lengths appears to be roughly symmetric, with no obvious outliers.'''

'''Plot 2: 
The scatter plot illustrates a positive relationship between 'sepal_length' and 'sepal_width' across different species of iris.
Setosa species generally have shorter but wider sepals compared to versicolor and virginica species.
Versicolor and virginica species show more overlap in sepal characteristics, with some versicolor specimens having similar sepal dimensions 
to virginica.'''


'''Plot 3: 
The boxplot demonstrates clear differences in 'petal_length' distributions among the three species of iris (setosa, versicolor, virginica).
Setosa species have significantly shorter petals compared to versicolor and virginica species.
Versicolor and virginica species exhibit a broader range of petal lengths, with virginica generally having longer petals than versicolor.'''

'''Plot 4: 
Petal measurements ('petal_length' and 'petal_width') appear to be highly correlated and effective in distinguishing between iris species.
Setosa species are well-separated from versicolor and virginica species based on petal characteristics.
Sepal measurements ('sepal_length' and 'sepal_width') show weaker correlations and less distinct separation between species compared 
to petal measurements.'''
```
