### Questin 01
Scientific Data Analysis with NumPy You are a scientist conducting research on climate change. You have collected temperature data from weather stations around the world stored in a NumPy array. Your task is to analyze the data using NumPy to:
1. Calculate the average temperature for each month across all stations.
2. Identify the station with the highest and lowest average temperature.
3. Determine the month with the largest temperature variation.
4. Visualize the temperature data using line plots or heatmaps.



<b>Use the following data set

```python
import numpy as np

# Sample temperature data for each month (in Celsius)
temperature_data = np.array([
    # January
    [10, 12, 15, 8, 9],
    # February
    [11, 13, 14, 9, 10],
    # March
    [12, 14, 16, 10, 11],
    # April
    [14, 16, 18, 12, 13],
    # May
    [16, 18, 20, 14, 15],
    # June
    [18, 20, 22, 16, 17],
    # July
    [20, 22, 24, 18, 19],
    # August
    [22, 24, 26, 20, 21],
    # September
    [20, 22, 24, 18, 19],
    # October
    [18, 20, 22, 16, 17],
    # November
    [15, 17, 19, 13, 14],
    # December
    [12, 14, 16, 10, 11]
])

# Station names for reference
station_names = ['Station A', 'Station B', 'Station C', 'Station D', 'Station E']
```

```python
mean_of_stations = np.mean(temperature_data,axis=0)
mean_of_stations
```

```python
maximum_temperature=np.max(mean_of_stations)
station_max=station_names[np.argmax(mean_of_stations)]
mini_temperature=np.min(mean_of_stations)
station_min=station_names[np.argmin(mean_of_stations)]
print(f"Station with the highest average teamperature is {station_max} with the temperature of ({maximum_temperature})")
print(f"Station with the highest average teamperature is {station_min} with the temperature of ({mini_temperature})")
```

```python
month_list=['Jan','Feb','March','April','May','June','July','August','Sept','Oct','Nov','Dec']
variance_temp=np.var(temperature_data,axis=1)
max_var=np.max(variance_temp)
month_with_max_var=month_list[np.argmax(variance_temp)]
print(f"Month with maximum varience is {month_with_max_var} with the varience of {max_var}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,10))
for i in range(len(station_names)):
    plt.plot(range(1, 13), temperature_data[:, i], label=station_names[i],color='blue')
plt.title("Temperature of All Stations")
plt.xlabel("Stations")
plt.ylabel("Temperature")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(temperature_data, cmap='magma', annot=True, fmt=".1f", xticklabels=station_names,
            yticklabels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title('Temperature Heatmap Across Stations and Months')
plt.xlabel('Station')
plt.ylabel('Month')
plt.show()
```

```python
grayscale_image_path = 'gray_image.jpg'
grayscale_image = cv2.imread(grayscale_image_path, cv2.IMREAD_GRAYSCALE)
threshold_value = 128

a, binary_image = cv2.threshold(grayscale_image, threshold_value, 255, cv2.THRESH_BINARY)
binary_image_path = 'bin_image.jpg'
cv2.imwrite(binary_image_path, binary_image)
```

```python
binary_image_path = 'bin_image.jpg'
binary_image = cv2.imread(binary_image_path, cv2.IMREAD_GRAYSCALE)
blurred_image = cv2.GaussianBlur(binary_image, (5, 5), 0)
blurred_image_path = 'blurred_bin_image.jpg'
cv2.imwrite(blurred_image_path, blurred_image)
```

```python
blurred_image_path = 'blurred_bin_image.jpg'
blurred_image = cv2.imread(blurred_image_path, cv2.IMREAD_GRAYSCALE)
sobelx = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=3)
magnitude = cv2.magnitude(sobelx, sobely)
magnitude = np.uint8(np.absolute(magnitude))
magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX)
edge_image_path = 'edge_detected_image.jpg'
cv2.imwrite(edge_image_path, magnitude)
```

```python
grayscale_image_path = 'gray_image.jpg'
grayscale_image = cv2.imread(grayscale_image_path, cv2.IMREAD_GRAYSCALE)
histogram = cv2.calcHist([grayscale_image], [0], None, [256], [0, 256])
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
```
